```python
# =============================
# Credit Card Fraud Detection (Decision Tree)
# Google Colab Ready
# 20 synthetic records, tabular visuals, diagrams, formulas + worked example
# =============================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from IPython.display import display, Markdown

# -----------------------------
# Step 0: Settings
# -----------------------------
np.random.seed(42)
pd.options.display.max_columns = None

# -----------------------------
# Step 1: Generate synthetic dataset (20 records)
# Features (X):
#  - transaction_amount (float, INR)
#  - time_of_day (morning/afternoon/evening/night)
#  - merchant_category (groceries/electronics/travel/restaurants)
#  - transaction_type (online/in-store)
#  - card_present (0/1)  # 1 if card physically present (chip/swipe), 0 otherwise (card-not-present)
#  - country_match (0/1) # 1 if billing country matches transaction country
# Label (Y):
#  - fraud: 'Fraud' or 'Legit'
# -----------------------------

n = 100 # Increased from 20 to 100
time_bins = ['morning', 'afternoon', 'evening', 'night']
merchant_cats = ['groceries', 'electronics', 'travel', 'restaurants']
transaction_types = ['online', 'in-store']

rows = []
# We'll bias fraudulent transactions to have characteristics like: high amount, online, card_not_present, country mismatch
for i in range(n):
    # pick base random
    amount = round(np.random.exponential(scale=80) + 10, 2)  # many small, some large
    tod = np.random.choice(time_bins)
    mcat = np.random.choice(merchant_cats)
    ttype = np.random.choice(transaction_types, p=[0.5, 0.5])
    card_present = 1 if ttype == 'in-store' and np.random.rand() > 0.2 else 0
    country_match = 1 if np.random.rand() > 0.15 else 0

    # Now decide label with some heuristics
    # Heuristic rules to set label (so model has learnable patterns)
    fraud_score = 0
    if amount > 300: fraud_score += 2
    if ttype == 'online': fraud_score += 1
    if card_present == 0: fraud_score += 1
    if country_match == 0: fraud_score += 1
    if mcat == 'electronics' and amount > 200: fraud_score += 1
    if tod == 'night' and ttype == 'online': fraud_score += 1

    label = 'Fraud' if fraud_score >= 3 else 'Legit'

    rows.append({
        'transaction_amount': amount,
        'time_of_day': tod,
        'merchant_category': mcat,
        'transaction_type': ttype,
        'card_present': card_present,
        'country_match': country_match,
        'label': label
    })

df = pd.DataFrame(rows).sample(frac=1, random_state=42).reset_index(drop=True)  # shuffle

# Save CSV for submission
csv_path = '/content/creditcard_fraud_100.csv' # Changed filename to reflect new size
df.to_csv(csv_path, index=False)

display(Markdown("## Step 1: Synthetic Dataset (100 records)")) # Changed display text
display(df)

print(f"\nCSV saved to: {csv_path}\n")

# -----------------------------
# Step 2: Prepare X and Y (encode categoricals)
# -----------------------------
display(Markdown("## Step 2: Prepare features (X) and label (Y)"))

X = df.drop('label', axis=1).copy()
y = df['label'].copy()

# Encode categorical features
encoders = {}
for col in ['time_of_day', 'merchant_category', 'transaction_type']:
    le = LabelEncoder()
    X[col + '_enc'] = le.fit_transform(X[col])
    encoders[col] = (le, dict(zip(le.classes_, le.transform(le.classes_))))

# Keep numeric & encoded columns; drop raw categoricals
X_model = X[['transaction_amount', 'time_of_day_enc', 'merchant_category_enc',
             'transaction_type_enc', 'card_present', 'country_match']].copy()

display(Markdown("### Encodings used:"))
enc_table = {k: v[1] for k, v in encoders.items()}
display(pd.DataFrame(enc_table).T)

display(Markdown("### X (features used for model):"))
display(X_model)

display(Markdown("### Y (labels):"))
display(y.to_frame('label'))

# -----------------------------
# Step 3: Train/Test split & Decision Tree
# -----------------------------
display(Markdown("## Step 3: Train Decision Tree"))

X_train, X_test, y_train, y_test = train_test_split(X_model, y, test_size=0.25, random_state=42, stratify=y)

dt = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=42)
dt.fit(X_train, y_train)

y_pred = dt.predict(X_test)

display(Markdown("### Performance on test set"))
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", round(acc, 3))
print("\nClassification Report:")
display(pd.DataFrame(classification_report(y_test, y_pred, output_dict=True)).transpose())

# Confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=['Legit', 'Fraud'])
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=['Legit','Fraud'], yticklabels=['Legit','Fraud'], cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# -----------------------------
# Step 4: Visualize Decision Tree & Feature Importance
# -----------------------------
display(Markdown("## Step 4: Decision Tree Visualization & Feature Importance"))

plt.figure(figsize=(14,6))
plot_tree(dt, feature_names=X_model.columns, class_names=dt.classes_, filled=True, rounded=True, fontsize=10)
plt.show()

fi = pd.Series(dt.feature_importances_, index=X_model.columns).sort_values(ascending=False)
display(Markdown("### Feature Importances"))
display(fi.to_frame('importance'))

plt.figure(figsize=(6,4))
fi.plot(kind='bar')
plt.title('Feature Importances')
plt.ylabel('Importance')
plt.tight_layout()
plt.show()

# -----------------------------
# Step 5: Predictions table (test set) - tabular visual
# -----------------------------
display(Markdown("## Step 5: Predictions Table (Test set)"))
test_results = X_test.copy()
test_results['actual'] = y_test.values
test_results['predicted'] = y_pred
# Show a readable table: decode encoded columns back for clarity
# invert encodings:
for col, (le, mapping) in encoders.items():
    enc_col = col + '_enc'
    inv_map = {v:k for k,v in mapping.items()}
    test_results[col] = test_results[enc_col].map(inv_map)
display(test_results[['transaction_amount','time_of_day','merchant_category','transaction_type','card_present','country_match','actual','predicted']])

# -----------------------------
# Step 6: Explanation in terms of X and Y + Decision Tree math
# -----------------------------
display(Markdown("## Step 6: Explanation (X and Y) & Decision Tree Math"))

display(Markdown("""
**1) Definitions**
- **X (features)**: the input vector for each transaction.
  In our project:
  \\[
  X = [\\text{transaction\_amount},\\ \\text{time\_of\_day\_enc},\\ \\text{merchant\_category\_enc},\\ \\text{transaction\_type\_enc},\\ \\text{card\_present},\\ \\text{country\_match}]
  \\]
- **Y (label)**: target variable we want to predict:
  \\[
  Y \\in \\{\\text{Legit},\\text{Fraud}\\}
  \\]

**2) Decision Tree Idea (brief)**
A Decision Tree splits data by choosing a feature and threshold that best separates classes at each node. We used **entropy** as the impurity measure and **information gain** to pick splits.

**3) Entropy formula**
For a node with class probabilities \\(p_1, p_2, ..., p_k\\):
\\[
H = -\\sum_{i=1}^k p_i \\log_2 p_i
\\]
For binary class (Legit/Fraud):
\\[
H = -p_{Legit} \\log_2 p_{Legit} - p_{Fraud} \\log_2 p_{Fraud}
\\]

**4) Information Gain (IG)**
If a node splits into left and right subsets, IG is:
\\[
IG = H_{parent} - \\left( \\frac{N_{left}}{N} H_{left} + \\frac{N_{right}}{N} H_{right} \\right)
\\]
The split with highest IG is selected.
"""))

# -----------------------------
# Step 7: Worked numerical example (compute entropy & info gain for a candidate split)
# We'll compute IG for splitting on "transaction_type_enc == online(??)" roughly:
# For demonstration pick split: transaction_type_enc == value_of('online')
# -----------------------------
display(Markdown("## Step 7: Worked Example â€” Entropy & Information Gain for a candidate split"))

# choose candidate feature: transaction_type_enc equality to the encoding of 'online'
le_tt, mapping_tt = encoders['transaction_type']
online_code = mapping_tt['online']  # numeric code for online

# Build parent distribution
parent_counts = y_train.value_counts()
N_parent = len(y_train)
p_parent = parent_counts / N_parent
def entropy_from_counts(counts):
    probs = counts / counts.sum()
    probs = probs[probs > 0]
    return -np.sum(probs * np.log2(probs))

H_parent = entropy_from_counts(parent_counts)
display(Markdown(f"Parent counts (train):\n{parent_counts.to_string()}"))
print("N_parent =", N_parent)
print("H_parent (bits) =", round(H_parent, 4))

# split by transaction_type_enc == online_code
mask_left = X_train['transaction_type_enc'] == online_code
left_counts = y_train[mask_left].value_counts()
right_counts = y_train[~mask_left].value_counts()
N_left = mask_left.sum()
N_right = (~mask_left).sum()

H_left = entropy_from_counts(left_counts) if N_left>0 else 0.0
H_right = entropy_from_counts(right_counts) if N_right>0 else 0.0

display(Markdown(f"Left (transaction_type == online) counts:\n{left_counts.to_string() if N_left>0 else 'none'}"))
display(Markdown(f"Right (transaction_type != online) counts:\n{right_counts.to_string() if N_right>0 else 'none'}"))

print("N_left =", N_left, " H_left =", round(H_left,4))
print("N_right =", N_right, " H_right =", round(H_right,4))

IG = H_parent - (N_left/N_parent)*H_left - (N_right/N_parent)*H_right
print("\nInformation Gain for split on transaction_type == 'online':", round(IG,4))

display(Markdown("**Interpretation:** A higher information gain means the split reduces uncertainty more. The tree selects the best feature & threshold using this idea at each node."))

# -----------------------------
# Step 8: Save model results & final notes
# -----------------------------
display(Markdown("## Step 8: Final Notes & Saved Files"))
display(Markdown(f"- Dataset (CSV): `{csv_path}`"))
display(Markdown("- Decision Tree trained with entropy, max_depth=3."))
display(Markdown("- You can explain X and Y like this: X are the input measurements for each transaction; Y is whether it is Fraud or Legit. The tree uses entropy & information gain to pick splits."))

# End of notebook cell.
```
